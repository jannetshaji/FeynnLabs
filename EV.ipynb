```python
x = df.iloc[:,[2,10]].values
print(x)
```

```python
wcss = []
for i in range(1,11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=25)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
```

```python
sns.set()
plt.plot(range(1,11), wcss)
plt.title('The Elbow Point Graph')
plt.xlabel('No. of Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans = KMeans(n_clusters=2, init='k-means++', random_state=0)
y = kmeans.fit_predict(x)
print(y)
```

```python
plt.figure(figsize=(10,10))
plt.scatter(x[y==0,0], x[y==0,1], s=100, c='green', label='Cluster 1')
plt.scatter(x[y==1,0], x[y==1,1], s=100, c='blue', label='Cluster 2')

plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=80, c='red', label='centroids')

plt.title('Vehicle Type')
plt.xlabel('Range')
plt.ylabel('Price_max')
plt.show()
```

```python
df = df.drop('ListColumn', axis=1)
```

```python
for label, content in df.items():
    if not pd.api.types.is_numeric_dtype(df[label]):
        lst = df[label].unique()
        label_map = {}
        for i in range(len(lst)):
            label_map[lst[i]] = i
        df[label] = df[label].map(label_map)
df.info()
```

```python
plt.figure(figsize=(10,5))
sns.heatmap(df.corr(), annot=True)
```

```python
df = df.drop('Transmission', axis=1)
```

```python
df = df.drop('VehicleType', axis=1)
```

```python
df = df.drop('Capacity', axis=1)
```

```python
plt.figure(figsize=(10,5))
sns.heatmap(df.corr(), annot=True)
```

```python
wcss = []
for k in range(1, 10):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(df)
    wcss.append(kmeans.inertia_)

plt.figure(figsize=(10, 5))
plt.plot(range(1, 10), wcss)
plt.xlabel('Number of clusters')
plt.ylabel('wcss')
plt.show()
```

```python
kmeans = KMeans(2, init = 'k-means++', random_state=42).fit(df)
df['cluster_number'] = kmeans.labels_
```

```python
df
```

```python
cross_state = pd.crosstab(df['cluster_number'], df['Style'])
cross_state
```

```python
plt.rcParams["figure.figsize"] = [200, 50]
plt.rcParams['font.size'] = 100
mosaic(cross_state.stack())
plt.show()
```

```python
cross_vehicleType = pd.crosstab(df['cluster_number'], df['TopModel'])
cross_vehicleType
```

```python
plt.rcParams["figure.figsize"] = [200, 50]
plt.rcParams['font.size'] = 100
mosaic(cross_vehicleType.stack(), horizontal=False)
plt.show()
```
